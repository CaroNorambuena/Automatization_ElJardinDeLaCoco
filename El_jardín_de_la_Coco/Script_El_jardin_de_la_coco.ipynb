1. Algoritmos para defir el precio de venta minorista considerando el coste de los materiales y el descuento del SII (IVA).

```python
from sympy import symbols, Eq, solve
```

```python
# Input: Variables que entrega el usuario
input1 = 2433
input2 = 3

# Ecuación 1
costomateriales = input1
ganancia = input2

valorneto = costomateriales * ganancia # Ec.1

# Output Ec.1
print(valorneto)
```

```python
## Sistema de Ecuaciones

# Definir variables
x, y = symbols('valorventa descIVA')

# Definir ecuaciones
eq2 = Eq(x/y , 1/0.19)
eq3 = Eq(x - y, valorneto)

# Resolver
sol = solve((eq2, eq3), (x, y)) #guarda las solucines en un diccionaro

print(sol)
```

```python
print(sol[x])
print(sol[y])
```

```python
valorventamin = sol[x]
print(valorventamin)
```

2. Algoritmos para definir el precio de venta mayorista

```python
## Modelos de disminución valor de venta vs cant de unidades

unidades = [unidad + 1 for unidad in range (4, 100)]

# Setear modelo
modelo = "lineal3"

if modelo == "lineal1":
    descuentos = []
    descuento = 0.05
    for i in range (100):
        descuento = round(descuento + 0.001, 3)
        descuentos.append(descuento)

if modelo == "lineal2":
    descuentos = []
    descuento = 0.05
    for i in range (100):
        descuento = round(descuento + 0.002, 3)
        descuentos.append(descuento)

if modelo == "lineal3":
    descuentos = []
    descuento = 0.05
    for i in range (100):
        descuento = round(descuento + 0.008, 3)
        descuentos.append(descuento)


porc_descuentos = [round(descuento * 100, 2) for descuento in descuentos]

descuentosventamay = [round(descuento*valorventamin) for descuento in descuentos]

valoresventamay = [round(valorventamin - descuentoventamay) for descuentoventamay in descuentosventamay]

valoresventamay_neto = [round(valorventamay - (valorventamay * 0.19)) for valorventamay in valoresventamay]

gananciasventamay = [round(valorventamay_neto - costomateriales) for valorventamay_neto in valoresventamay_neto]

porc_gananciasventamay = [round((gananciaventamay * 100)/costomateriales) for gananciaventamay in gananciasventamay]

# Combinar las listas para formar tuplas. Nota: Los datos en tuplas son facilmente transformables en Dataframe
datos = list(zip(unidades, descuentos, porc_descuentos, descuentosventamay, valoresventamay, valoresventamay_neto, gananciasventamay, porc_gananciasventamay))
print(datos)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(datos, columns=["Unidades", "Descuento", "Descuento %", "Valor descuento", "Valor venta mayorista", "Valor neto venta mayorista", "Ganancia", "% Ganancia"])
print(df)
```

```python
df.to_excel("Tabla valores venta mayorista sustrato.xlsx", index=False)
```

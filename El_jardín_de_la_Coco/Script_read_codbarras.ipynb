# Lector de códigos de barra con el celular

Cuaderno que contiene la automatización de la lectura de códigos de barra mediante la cámara del celular. Para ello se ocupa IP Webcam que media la comunicación entre la camara del celular y el PC. Es necesario la conexión a una red.

Generador de códigos de barra

Para seguimiento logístico. Un código universal se debe pedir a https://gs1chile.org/Codigos

```python
import barcode
from barcode.writer import ImageWriter

CODABAR = barcode.get_barcode_class('codabar')

for i in range (1): # seteado en "1" para que itere solo 1 vez. El output será "A0A". Nota: range(i, j) produces i, i+1, i+2, ..., j-1
 serie = "A" + str(i) + "A"
 codigo = CODABAR(serie, writer=ImageWriter())
 codigo.save(serie) #si el script se ejecuta en una carpeta específica los códigos quedan guardados en esa carpeta
```
